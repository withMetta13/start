```python
import openpyxl
wb = openpyxl.Workbook()
sheet = wb.active
for i in range(1, 11): # create some data in column A
     sheet['A' + str(i)] = i

refObj = openpyxl.chart.Reference(sheet, min_col=1, min_row=1, max_col=1,max_row=10)
seriesObj = openpyxl.chart.Series(refObj, title='First series')

chartObj = openpyxl.chart.BarChart()
chartObj.title = 'My Chart'
chartObj.append(seriesObj)

sheet.add_chart(chartObj, 'C5')
wb.save('sampleChart.xlsx')
```

```python
def multi(num):
    wb = openpyxl.Workbook()
    sheet = wb.active
    for i in range(num):
        sheet.cell(row = 1, column =i+2).value = i+1
        sheet.cell(row = i+2, column =1).value = i+1
    for r in range(2, num+2):
        for c in range(2, num+2):
            sheet.cell(row = r, column= c).value = (r-1) * (c-1)
    
    wb.save('example1.xlsx')
    
multi(6)
```

```python
wb = openpyxl.load_workbook('example.xlsx')
sheet = wb.active
data=[]
for r in range(sheet.max_row):
    r_data = []
    for c in range(sheet.max_column):
        r_data.append(sheet.cell(row = r+1, column=c+1).value)
    data.append(r_data)

wb1 = openpyxl.Workbook()
sheet1 = wb1.active
for r in range(sheet.max_column):
    for c in range(sheet.max_row):
        sheet1.cell(row = r+1, column=c+1).value = data[c][r]    
wb1.save('example3.xlsx')
```
